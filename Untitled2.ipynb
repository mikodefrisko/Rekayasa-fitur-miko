```python
import pandas as pd
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns
from google_play_scraper import Sort, reviews
from transformers import pipeline

# ==========================================
# 1. SCRAPING DATA (BPJS MOBILE)
# ==========================================
APP_ID = 'app.bpjs.mobile'
COUNT = 500

print(f"🔄 1. Sedang mengambil {COUNT} komentar dari {APP_ID}...")

try:
    result, continuation_token = reviews(
        APP_ID,
        lang='id', country='id', sort=Sort.NEWEST, count=COUNT
    )

    df = pd.DataFrame(result)
    df = df[['content', 'score', 'at']]
    df.rename(columns={'content': 'text', 'score': 'rating', 'at': 'date'}, inplace=True)
    print(f"   ✅ Berhasil mengambil {len(df)} data.")

    # ==========================================
    # 2. PREPROCESSING
    # ==========================================
    print("🔄 2. Sedang membersihkan teks...")

    def clean_text(text):
        text = str(text).lower()
        text = re.sub(r'\d+', '', text)
        text = text.translate(str.maketrans('', '', string.punctuation))
        text = text.strip()
        return text

    df['clean_text'] = df['text'].apply(clean_text)
    df = df[df['clean_text'] != '']
    print("   ✅ Pembersihan selesai.")

    # ==========================================
    # 3. ANALISIS SENTIMEN (NLP)
    # ==========================================
    print("🔄 3. Sedang memuat Model NLP (Tunggu 1-2 menit)...")
    print("   (Jika error 'OSError', tolong jalankan ulang cell ini!)")

    # Load model
    sentiment_analysis = pipeline("sentiment-analysis", model="w11wo/indonesian-roberta-base-sentiment-classifier-01")

    def get_sentiment_nlp(text):
        try:
            # Batasi teks max 512 karakter
            return sentiment_analysis(text[:512])[0]['label']
        except:
            return "neutral"

    print("   Sedang memproses sentimen satu per satu...")
    df['sentiment_pred'] = df['clean_text'].apply(get_sentiment_nlp)
    print("   ✅ Analisis sentimen selesai!")

    # ==========================================
    # 4. VISUALISASI & SIMPAN
    # ==========================================
    print("🔄 4. Membuat grafik...")

    # Grafik
    plt.figure(figsize=(8, 5))
    sns.countplot(x='sentiment_pred', data=df, palette='viridis')
    plt.title(f'Distribusi Sentimen {APP_ID}')
    plt.show()

    # Simpan CSV
    df.to_csv('tugas_selesai.csv', index=False)
    print("\n🎉 SELESAI SEMUA! File 'tugas_selesai.csv' siap didownload di folder kiri.")
    print("Contoh Data:")
    print(df[['clean_text', 'sentiment_pred']].head())

except Exception as e:
    print(f"\n❌ TERJADI ERROR: {e}")
    print("TIPS: Jika errornya 'OSError' atau 'huggingface', coba tekan tombol Play ▶️ sekali lagi.")
```

```python
import pandas as pd
import re
import string
import matplotlib.pyplot as plt
import seaborn as sns
from google_play_scraper import Sort, reviews
from transformers import pipeline

# --- KONFIGURASI ---
APP_ID = 'app.bpjs.mobile'  # Kita pakai BPJS karena PLN susah discrape
COUNT = 500

print(f"🚀 MEMULAI PROSES UNTUK APLIKASI: {APP_ID}")

# ==========================================
# 1. SCRAPING
# ==========================================
print("\n[1/4] Sedang mengambil data ulasan...")
try:
    result, _ = reviews(
        APP_ID,
        lang='id', country='id', sort=Sort.NEWEST, count=COUNT
    )

    if len(result) == 0:
        raise ValueError("Tidak ada data yang terambil. Coba cek koneksi internet.")

    df = pd.DataFrame(result)
    df = df[['content', 'score', 'at']]
    df.rename(columns={'content': 'text', 'score': 'rating', 'at': 'date'}, inplace=True)
    print(f"   ✅ Berhasil mengambil {len(df)} ulasan.")

except Exception as e:
    print(f"   ❌ Gagal Scraping: {e}")
    # Stop proses jika scraping gagal
    df = pd.DataFrame()

if not df.empty:
    # ==========================================
    # 2. PREPROCESSING
    # ==========================================
    print("\n[2/4] Membersihkan teks...")

    def clean_text(text):
        text = str(text).lower()
        text = re.sub(r'\d+', '', text)
        text = text.translate(str.maketrans('', '', string.punctuation))
        return text.strip()

    df['clean_text'] = df['text'].apply(clean_text)
    df = df[df['clean_text'] != '']
    print("   ✅ Pembersihan selesai.")

    # ==========================================
    # 3. ANALISIS SENTIMEN (DENGAN CADANGAN)
    # ==========================================
    print("\n[3/4] Memuat Model AI (Ini butuh internet stabil)...")

    model_name = "w11wo/indonesian-roberta-base-sentiment-classifier-01"
    fallback_model = "mdhugol/indonesia-bert-sentiment-classification"

    try:
        # Coba Model Utama
        print(f"   Mencoba download model utama: {model_name}")
        sentiment_analysis = pipeline("sentiment-analysis", model=model_name)
    except Exception as e:
        print(f"   ⚠️ Model utama gagal ({e}). Mencoba model cadangan...")
        try:
            # Coba Model Cadangan jika utama gagal
            sentiment_analysis = pipeline("sentiment-analysis", model=fallback_model)
        except:
            print("   ❌ Gagal memuat semua model. Pastikan internet Colab lancar.")
            sentiment_analysis = None

    if sentiment_analysis:
        print("   ✅ Model berhasil dimuat! Sedang menganalisis...")

        def get_sentiment(text):
            try:
                # Potong teks max 512 huruf
                return sentiment_analysis(text[:512])[0]['label']
            except:
                return "neutral"

        df['sentiment_pred'] = df['clean_text'].apply(get_sentiment)
        print("   ✅ Analisis sentimen selesai.")

        # ==========================================
        # 4. VISUALISASI & SIMPAN
        # ==========================================
        print("\n[4/4] Membuat Laporan...")

        # Grafik
        try:
            plt.figure(figsize=(8, 5))
            sns.countplot(x='sentiment_pred', data=df, palette='viridis')
            plt.title(f'Sentimen Analisis: {APP_ID}')
            plt.show()
        except:
            print("   (Grafik tidak bisa ditampilkan, tapi data aman)")

        # Simpan
        filename = 'tugas_selesai_bpjs.csv'
        df.to_csv(filename, index=False)
        print(f"\n🎉 SELESAI! File '{filename}' sudah jadi. Cek folder di kiri layar.")
        print(df[['clean_text', 'sentiment_pred']].head())

    else:
        print("\n❌ Proses berhenti karena Model AI gagal dimuat.")

else:
    print("\n❌ Proses berhenti karena Scraping gagal.")
```
